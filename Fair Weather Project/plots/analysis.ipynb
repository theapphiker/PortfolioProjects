```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
# load old weather dataset
df1=pd.read_csv(r'C:\Users\syast\OneDrive\Documents\Data Analysis Practice\Fair Weather Project\data\old_weather_cleaned.csv')
```

```python
# load new weather dataset
df2=pd.read_csv(r'C:\Users\syast\OneDrive\Documents\Data Analysis Practice\Fair Weather Project\data\new_weather_cleaned.csv')
```

```python
# print mean statistics about old temperatures and new temperatures
print('The mean max temperature during fair week from 1960 to 1980: %d degrees Farenheit' % df1['TMAX'].mean())
print('The mean min temperature during fair week from 1960 to 1980: %d degrees Farenheit' % df1['TMIN'].mean())
print('The mean max temperature during fair week from 2013 to 2017: %d degrees Farenheit' % df2['TMAX'].mean())
print('The mean min temperature during fair week from 2013 to 2017: %d degrees Farenheit' % df2['TMIN'].mean())
```

```python
# print basic statistics about old temperatures
df1[['TMAX','TMIN']].describe()
```

```python
# print basic statistics about new temperatures
df2[['TMAX','TMIN']].describe()
```

```python
# print median statistics about old and new temperatures
print('The median max temperature during fair week from 1960 to 1980: %d degrees Farenheit' % df1['TMAX'].median())
print('The median min temperature during fair week from 1960 to 1980: %d degrees Farenheit' % df1['TMIN'].median())
print('The median max temperature during fair week from 2013 to 2017: %d degrees Farenheit' % df2['TMAX'].median())
print('The median min temperature during fair week from 2013 to 2017: %d degrees Farenheit' % df2['TMIN'].median())
```

```python
# convert column to datetime
df1['DATE']=pd.to_datetime(df1['DATE'])
df2['DATE']=pd.to_datetime(df2['DATE'])
```

```python
# plot scatter plot showing max temperatures
plt.figure(figsize=(15, 7))
plt.scatter(df1['DATE'],df1['TMAX'])
plt.scatter(df2['DATE'],df2['TMAX'])
plt.grid()
plt.title('Scatter Plot showing Max Temperatures for Bloomsburg Fair')
plt.xlabel('Years')
plt.ylabel('Temperature in Farenheit')
plt.show()
```

```python
# plot scatter plot showing max temperatures
plt.figure(figsize=(15, 7))
plt.scatter(df1['DATE'],df1['TMIN'])
plt.scatter(df2['DATE'],df2['TMIN'])
plt.grid()
plt.title('Scatter Plot showing Min Temperatures for Bloomsburg Fair')
plt.xlabel('Years')
plt.ylabel('Temperature in Farenheit')
plt.show()
```

```python
# plot box plot showing min temperatures
fig=plt.figure(figsize=(15, 7))
ax=fig.add_subplot(111)
Temps_1960_1980=df1['TMIN']
Temps_2013_2017=df2['TMIN']
data=[Temps_1960_1980,Temps_2013_2017]
bp=ax.boxplot(data)
plt.grid()
plt.title('Box Plot showing Min Temperatures for Bloomsburg Fair')
ax.set_xticklabels(['Min Temp 1960-1980', 'Min Temp 2013-2017'])
plt.ylabel('Temperature in Farenheit')
plt.show()
```

```python
# plot box plot showing min temperatures
fig=plt.figure(figsize=(15, 7))
ax=fig.add_subplot(111)
Temps_1960_1980=df1['TMAX']
Temps_2013_2017=df2['TMAX']
data=[Temps_1960_1980,Temps_2013_2017]
bp=ax.boxplot(data)
plt.grid()
plt.title('Box Plot showing Max Temperatures for Bloomsburg Fair')
ax.set_xticklabels(['Max Temp 1960-1980', 'Max Temp 2013-2017'])
plt.ylabel('Temperature in Farenheit')
plt.show()
```
